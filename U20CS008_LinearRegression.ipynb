```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris, make_regression
```

```python
iris = load_iris()
data = iris['data']
feature_names = iris['feature_names']

Y = data[:, 0:1]
Y_feature_name = feature_names[0]

X = data[:, 1:4]
X_feature_name = feature_names[1:4]

x_train, x_test, y_train, y_test = train_test_split(X, Y, random_state=0)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

df = pd.DataFrame(data=data, index=list(range(0, data.shape[0])), columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width']);
# print(df.head())

vif_data = pd.DataFrame()
vif_data["feature"] = feature_names[1:]
# print(X)
vif_data["VIF"] = [variance_inflation_factor(X, i) for i in range(X.shape[1])]
  
print(vif_data)
```

```python
for i in range(0, 3):
  linear_regression = LinearRegression().fit(x_train[:, i:i+1], y_train)

  plt.figure()
  plt.scatter(X[:, i], Y[:, 0], c="gray")
  plt.xlabel(feature_names[i+1])
  plt.ylabel('sepal length (cm)')
  
  m = linear_regression.coef_[0][0]
  c = linear_regression.intercept_[0]

  print("coefficient of correlation: " + str(m))

  x_line = np.linspace(0, 8, 112)
  y_line = m * x_line + c

  plt.plot(x_line, y_line, "b")
  plt.show()
```

```python
linear_regression = LinearRegression().fit(x_train[:, 0:3], y_train)
  
m1, m2, m3 = linear_regression.coef_[0][0], linear_regression.coef_[0][1], linear_regression.coef_[0][2]
c = linear_regression.intercept_[0]

print("coefficient of correlation: " + str(linear_regression.coef_))

print("accuracy score: " + str(linear_regression.score(x_test, y_test)))
```
